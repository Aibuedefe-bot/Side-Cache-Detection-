```python
import pandas as pd
import os
```

```python
# Folder where you unzipped the files
DATA_FOLDER = r"C:\Users\edecr\Downloads\dataverse_files"
OUTPUT_FILE = "combined_harpy.csv"
```

```python
# All files and their labels (0=benign, 1=attack)
files = {
    "prog_basic_perf_w_instructions.csv": 0,
    "stress_c_basic_perf_w_instructions.csv": 0,
    "stress_m_basic_perf_w_instructions.csv": 0,
    "mb_bitcount_basic_perf_w_instructions.csv": 0,
    "stream_basic_perf_w_instructions.csv": 0,
    "bzip2_basic_perf_w_instructions.csv": 0,
    "ffmpeg_basic_perf_w_instructions.csv": 0,
    
    "meltdown_basic_perf_w_instructions.csv": 1,
    "spectre_v1_basic_perf_w_instructions.csv": 1,
    "spectre_v2_basic_perf_w_instructions.csv": 1,
    "spectre_v4_basic_perf_w_instructions.csv": 1,
    "zombieload_basic_perf_w_instructions.csv": 1,
    "fallout_basic_perf_w_instructions.csv": 1,
    "crosstalk_basic_perf_w_instructions.csv": 1,
}
```

```python
all_data = []
```

```python
print("Loading and labeling files...")
for filename, label in files.items():
    path = os.path.join(DATA_FOLDER, filename)
    if os.path.exists(path):
        df = pd.read_csv(path)
        df['label'] = label
        df['program'] = filename.replace("_basic_perf_w_instructions.csv", "")
        all_data.append(df)
        print(f"✓ {filename} → {len(df)} rows, label={label}")
    else:
        print(f"⚠️ Missing: {filename}")

# Combine everything
combined = pd.concat(all_data, ignore_index=True)

# Clean up column names (make them nicer)
combined.columns = [col.strip() for col in combined.columns]

print(f"\n✅ Combined dataset ready!")
print(f"   Total rows: {len(combined):,}")
print(f"   Benign: {len(combined[combined['label']==0]):,}")
print(f"   Attack: {len(combined[combined['label']==1]):,}")
print(f"   Columns: {list(combined.columns)}")

# Save it
combined.to_csv(OUTPUT_FILE, index=False)
print(f"\n💾 Saved as {OUTPUT_FILE} — ready for feature engineering!")
```
